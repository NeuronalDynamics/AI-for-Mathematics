## Generate Parenthesizations

Generate all possible parenthesizations of length $n$ for $1 \le n \le 10$. Files should be outputted to `data/catalan_parenthesizations_n.txt` and a plot of the time used to generate each data file should go in `figures/catalan_parenthesizations.png`.

```python
for n in range(1, 11):
  print(f"Parenthesizations for n={n} and size= {len(catalan.parenthesizations(n))}: {catalan.parenthesizations(n)}")
```

```python
# Dictionary to store the time taken for each n
time_taken = {}

# Generate parenthesizations for n from 1 to 10 and save to files
for n in range(1, 11):
    start_time = time.time()
    parenthesizations_n = catalan.parenthesizations(n)
    end_time = time.time()
    
    # Record the time taken
    time_taken[n] = end_time - start_time
    
    # Write the parenthesizations to a file
    with open(f'data/catalan_parenthesizations_{n}.txt', 'w') as file:
        for p in parenthesizations_n:
            file.write(p + '\n')

# Plot the time taken
plt.figure(figsize=(10, 6))
plt.plot(list(time_taken.keys()), list(time_taken.values()), marker='o')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Time Taken to Generate Parenthesizations')
plt.savefig('figures/catalan_parenthesizations.png')
plt.show()
```
